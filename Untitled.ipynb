```python
# Регрессия - набор факторов => конкретное значение
# 28 дней назад => 7 дней вперед
```

```python
import pandas as pd
import sklearn
```

```python
money = pd.read_csv("usd_rate.csv", delimiter = ";")
```

```python
money
```

```python
values = money["curs"]
```

```python
past = 7 * 4
future = 7
```

```python
start = past
end = len(values) - future
print(start, end)
```

```python
raw_df = []
for i in range(start, end):
    past_and_future_val = values[(i-past):(i+future)] # -28...+7
    raw_df.append(list(past_and_future_val))
```

```python
past_columns = [f"past_{i}" for i in range(past)]
```

```python
future_columns = [f"future_{i}" for i in range(future)]
```

```python
df = pd.DataFrame(raw_df, columns = (past_columns + future_columns))
```

```python
# Тренировочная выборка
x = df[past_columns][:-1] # Факторы, по которым делается предсказание
y = df[future_columns][:-1] # Значения, которые собираемся предсказать
# Мульти-регрессия - предсказываем сразу 7 значений
```

```python
# Тестовая выборка
x_test = df[past_columns][-1:]
y_test = df[future_columns][-1:]
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
plt.plot(values)
plt.show()
```

```python
LinReg = LinearRegression()
```

```python
LinReg.fit(x, y) # Обучение
```

```python
prediction = LinReg.predict(x_test) # Предсказание
```

```python
prediction[0]
```

```python
plt.plot(prediction[0])
plt.show()
```

```python
plt.plot(y_test.iloc[0])
plt.show()
```

```python
plt.plot(prediction[0], label = "prediction")
plt.plot(y_test.iloc[0], label = "real")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# средняя абсолютная ошибка, средняя квадратичная
```

```python
mean_absolute_error(y_test, prediction)
```

```python
mean_squared_error(y_test, prediction)
```

```python
MLP.fit(x, y)
```

```python
prediction = MLP.predict(x_test) # Предсказание
```

```python
plt.plot(prediction[0], label = "prediction")
plt.plot(y_test.iloc[0], label = "real")
plt.show()
```

```python
print("mean_absolute_error", mean_absolute_error(y_test, prediction))
print("mean_squared_error", mean_squared_error(y_test, prediction))
```

```python
#GridSearch
from sklearn.model_selection import GridSearchCV
```

```python
# Grid - набор параметров, среди которых мы ищем оптимальные
# Search - поиск
# CV - cross validation = кросс-валидация (обучение на разных частях данных)

# [X X X X * * * * * * * *]
# [* * * * X X X X * * * *]
# [* * * * * * * * X X X X]
```

```python
MLP2 = MLPRegressor()
grid_search = GridSearchCV(MLP2, {
    'activation': ['tanh', 'relu', 'logistic'],
    'alpha': linspace(0.1, 0.000001, 10),
    'hidden_layer_sizes': [(50,50,50), (100,50,25)],
    'max_iter': [200, 515, 1010, 2000],
    'random_state': [100, 150, 200],
    'learning_rate': ['adaptive', 'invscaling']
})
```

```python
MLP2
```

```python
grid_search.fit(x, y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.predict(x_test)
```

```python
prediction = grid_search.best_estimator_.predict(x_test)
plt.plot(prediction[0], label = "prediction")
plt.plot(y_test.iloc[0], label = "real")
plt.show()
```
